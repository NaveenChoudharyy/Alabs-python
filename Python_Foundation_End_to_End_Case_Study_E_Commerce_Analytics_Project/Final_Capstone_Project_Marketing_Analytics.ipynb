### Understanding how many new customers acquired every month

```python
order.info()
```

```python
## If you are ordering for the first time you are new customer

order['YM'] = order['order_purchase_timestamp'].dt.strftime('%Y, %m')


cust_order = pd.merge(order, cust, on='customer_id', how='left')


first_visit = cust_order.drop_duplicates(subset='customer_unique_id', keep='first')


first_visit_monthwise = cust_order[['YM', 'customer_unique_id']].drop_duplicates(subset='customer_unique_id', keep='first')


unique_cust_count_first_visit = first_visit_monthwise.groupby(by='YM')[['customer_unique_id']].count()



# Chart

plt.figure(figsize=(15,6))

plt.ticklabel_format(style='plain')
sns.lineplot(x=unique_cust_count_first_visit.index, y= unique_cust_count_first_visit.customer_unique_id, color='red')
plt.title('Monthly treand of new customers',pad=30,fontsize =24)
plt.ylabel('Count of customers',fontsize =20)
plt.xlabel('Months',fontsize =20)
plt.xticks(rotation=50)
plt.grid(alpha=0.4)

plt.tight_layout()
plt.show()
```

### Understand the retention of customers on month on month basis

```python
# If not ordering for first time, you are retained customer for that month

retained_cust = cust_order[~cust_order['customer_id'].isin(first_visit['customer_id'])]


retained_cust_monthwise = retained_cust.groupby(by='YM')[['customer_unique_id']].nunique()


# Chart

plt.figure(figsize=(15,6))

plt.ticklabel_format(style='plain')
sns.lineplot(x=retained_cust_monthwise.index, y= retained_cust_monthwise.customer_unique_id, color='red')
plt.title('Monthly treand of retaind customers',pad=30,fontsize =24)
plt.ylabel('Count of customers',fontsize =20)
plt.xlabel('Months',fontsize =20)
plt.xticks(rotation=40)
plt.grid(alpha=0.4)

plt.tight_layout()
plt.show()
```

## Performing Customers/sellers Segmentation

### Divide the customers into groups based on the revenue generated 

```python
D1 = pd.merge(cust,order, on = 'customer_id', how= 'right')


D2 = pd.merge(payment, D1, on ='order_id', 
              how='right')[['payment_value', 
                           'customer_unique_id']].fillna(value=0)


D3 = D2.groupby(['customer_unique_id'])[['payment_value']].sum().reset_index()


bins_D1 = pd.qcut(D3['payment_value'], q=10, duplicates='drop', labels=False)
D4 = D3.groupby(by=bins_D1)[['customer_unique_id']].count().reset_index()

D5 = D3.copy()
D5['bin']= pd.qcut(D3['payment_value'], q=10, duplicates='drop', labels=False)
D5['bin_2']= pd.qcut(D3['payment_value'], q=10, duplicates='drop')
print(D5.head(10))

# Chart
plt.figure(figsize=(15,6))
sns.barplot(y= D4.payment_value, x=D4.customer_unique_id, orient='h')
plt.title('Count of Customers Grouped by Revenue', fontsize=25, pad=30)
plt.ylabel('Revanue bins',fontsize=20)
plt.xlabel('Number of customers',fontsize=18)

plt.grid(linestyle=':', color= 'lightblue',alpha=.4,gapcolor='black', axis='x')
plt.tight_layout()
plt.show()
```

### Divide the sellers into groups based on the revenue generated 

```python
E1 = pd.merge(order_item, seller, on='seller_id',
              how='inner').merge(order, on='order_id', 
                                 how='inner').merge(payment, on='order_id', how='left')


E2 = E1.drop_duplicates(subset='order_id')[['seller_id', 'payment_value']].fillna(value=0)


E3 = E2.groupby(['seller_id'])[['payment_value']].sum().reset_index()


bins_E1 = pd.qcut(E2.payment_value, q=10, duplicates='drop', labels=False)


E4 = E3.groupby(by=bins_E1)[['seller_id']].count().reset_index()


E5 = E3.copy()
E5['bin']= pd.qcut(E3['payment_value'], q=10, duplicates='drop', labels=False)
E5['bin2']= pd.qcut(E3['payment_value'], q=10, duplicates='drop')
print(E5.head(10))

# Chart
plt.figure(figsize=(15,6))
sns.barplot(y= E4.payment_value, x=E4.seller_id, orient='h')
plt.title('Count of Sellers Grouped by Revenue', fontsize=25, pad=30)
plt.ylabel('Revanue bins',fontsize=20)
plt.xlabel('Number of sellers',fontsize=18)

plt.grid(linestyle=':', color= 'lightblue',alpha=.4,gapcolor='black', axis='x')
plt.tight_layout()
plt.show()
```

## Cross-Selling (Which products are selling together)
Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)


```python
F1 = pd.merge(order, order_item, on='order_id', how='inner').merge(cust, on='customer_id', how='left')


F2 = F1[['order_id', 'product_id', 'seller_id', 'customer_unique_id','order_purchase_timestamp']].drop_duplicates()


F3 = pd.merge(F2,F2, on =['order_id','seller_id','customer_unique_id',
                          'order_purchase_timestamp'])[['product_id_x','product_id_y','order_id','seller_id',
                                                        'customer_unique_id','order_purchase_timestamp']]


F4 = F3[F3['product_id_x'] != F3['product_id_y']]


F5 = pd.DataFrame({'col':F4['product_id_x'] + ' and ' + F4['product_id_y']})


F6 = F5.groupby(by='col')[['col']].count()


F7 = F6.rename(columns={'col':'count'}).reset_index().sort_values(by='count', ascending=False)


print(F7.head(10))
```
